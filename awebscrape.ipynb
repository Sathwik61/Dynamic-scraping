```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from time import sleep

# Path to the ChromeDriver executable
chrome_driver_path = r'C:\Users\SATHWIK\Downloads\chromedriver-win64 (1)\chromedriver-win64\chromedriver.exe'

# Website url
url = 'https://amazon.in'

# Set up Chrome WebDriver with service
service = Service(executable_path=chrome_driver_path)
driver = webdriver.Chrome(service=service)

# Navigate to a webpage
driver.get(url)

# Find search input and search button
input_search = driver.find_element("id",'twotabsearchtextbox')
search_btn = driver.find_element("id", "nav-search-submit-button")
# find_element(By.XPATH, '//input[@type='submit'])[1]')
# Input search query
input_search.send_keys("Smartphone under 10000")
sleep(1)

# Click search button
search_btn.click()
```

```python
from selenium.webdriver.common.by import By
products=[]
# title='a-size-medium a-color-base a-text-normal'
price=[]
title=[]
item=driver.find_elements(By.XPATH,"//span[@class='a-size-medium a-color-base a-text-normal']")
price_of_item=driver.find_elements(By.XPATH,"//span[@class='a-price-whole']")

for i in item:
    title.append(i.text)

    
for i in price_of_item:
    price.append(i.text.replace(",",""))
    
    
for t,p in zip(title,price):
    products.append({"Title":t,"price":p})
    
# print((products))
# print("\n\n\n**************************\n\n")
# print(price)
    
# print(len(products))
# print(len(price))


```
